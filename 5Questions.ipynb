#####  Given a passenger embarked on SouthHampton, what is the probability that he/she is young <12

```python
q1df = pd.read_csv('titanic.csv')
q1df = q1df.dropna(subset=['Age'])
q1df['young'] = q1df['Age']<12
```

```python
q1df.groupby(['Embarked','young']).count()
```

We see that the probability is 54/554 ~ 9.7%

##### Given that a passenger paid nothing, what is the probability that he/she survived?


```python
q2df = pd.read_csv('titanic.csv')
q2df['free'] = q2df['Fare'] == 0
```

```python
q2df.groupby(['free','Survived']).count()
```

We see that the probability is 1/15 ~ 6.7%

##### Given that a passenger is male and old > 65, what is the probability that he survived?


```python
q3df = pd.read_csv('titanic.csv')
q3df = q3df.dropna(subset=['Age'])
q3df['oldMale'] = (q3df['Age'] > 65) & (q3df['Sex'] == 'male')
```

```python
q3df.groupby(['oldMale','Survived']).count()
```

We see that the probability is 1/8 ~ 12.5%

##### Given that someone was a minor who survived, how likely was he/she to be a female?


```python
q4df = pd.read_csv('titanic.csv')
q4df = q4df.dropna(subset=['Age'])
q4df['minorSurvived'] = (q4df['Age']<18) & (q4df['Survived'] == 1)
```

```python
q4df.groupby(['minorSurvived','Sex']).count()
```

We see that the probability is 38/61 ~ 62.3%

##### Given that a passenger is under 30 but over 20 years old, what are the chances they are in first class?

```python
q5df = pd.read_csv('titanic.csv')
q5df = q5df.dropna(subset=['Age'])
q5df['20to30'] = (q5df['Age']<30) & (q5df['Age']>20)
```

```python
q5df.groupby(['20to30','Pclass']).count()
```

We see that the probability is 34/205 ~ 16.6%
